# Iterator:

* Iterator is an object that allow the programmer to traverse true a sequence of data without having to store them enter data in the memory.

*  An Iterator lets you fetch item one by one

* It does not store entire data in the memory it generates one at a time 

* Once an iterator is exhausted, it can't be reused

# Implementing and creating iterator object:

# We use two types :

1.iter(): return iterator object

2.next(): fetch item one at a time.

```python
l1 = [10, 20, 'hii', 'false', 5 - 3j, 3.89]             # 1 way 
iter_obj = iter(l1)
print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
```

```python
l1 = [10, 20, 'hii', 'false', 5 - 3j, 3.89]           # 2 way 
iter_obj = iter(l1)
print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
for i in iter_obj:
    print(i)
```

# Difference b/w 

* Iterator: Iterator is an object the actual give the value one by one


* Iteration: Process og going through items one by one .


* Iterable: An Iterable is something which can loop over.


```python
# Defining our own loop:

def myloop(iteratable):
    iterator = iter(iteratable)

    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break


myloop([10, 20, 30])
myloop(['hii', False, 3.44])
myloop({'a': 10, 'b': 20, 'c': 30})
myloop("hii")
```

```python
class MyRange:

    def __init__(self, num):
        self.num = num
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num < self.current:
            raise StopIteration
        val = self.current
        self.current += 1
        return val


obj = MyRange(5)
for i in obj:
    print(i)    # creating our one range function
```

# Generator:

* Simple way of creating iterator.

* A generator is special type of function that produces values one at a time by using yield keyword instance of returning all the value at once.

#  use of yield keyword:
* It pauses execution remember its state and resumes when the next value requested.

# Syntax:

* def func_name():
    * yield val1
     * yield val2
    *  | 
    *  |
    *  |
    * yield valn
* obj = func_name()
* print(next(obj))
* print(next(obj))


```python
def demo():
    yield "first line"
    yield "second line"
    yield "third line"

val = demo()
print(next(val))
print(next(val))
print(next(val))

```

```python
def infinite_numbers():
    i = 1
    while True:
        yield i
        i += 1


numbers = infinite_numbers()   

print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
```

# what is difference b/w normal function and generator.

| Normal Function                 | Generator Function                |
| ------------------------------- | --------------------------------- |
| Uses `return`                   | Uses `yield`                      |
| Returns complete result at once | Returns values one by one         |
| Stops after `return`            | Pauses at `yield`, resumes later  |
| Stores full data in memory      | Memory efficient (lazy)           |
| Cannot remember state           | Remembers last state              |
| Example: returns list           | Example: returns generator object |



# How ways to create a generator

* Using yield-

* Generator Expression (like list comprehension)

g = (i*i for i in range(5))


# what is the use of yield keyword.

* Converts a function into a generator

* Returns value without stopping the function

* Pauses execution and remembers state

* Produces values one at a time

| `yield`                  | `return`                 |
| ------------------------ | ------------------------ |
| Pauses function          | Ends function            |

# 5️⃣ What is an Iterator?

An iterator is an object that:

* Has __iter__() and __next__() methods

* Produces values one by one

* Raises StopIteration when finished

```python
l = [1,2,3]
it = iter(l)

print(next(it))
```

# 6️⃣ Difference between Iterable and Iterator. 

| Iterable                              | Iterator                                   |
|---------------------------------------|---------------------------------------------|
| Can be looped over                   | Actually produces values one by one        |
| Has `__iter__()` method              | Has both `__iter__()` and `__next__()`     |
| Does not track current position      | Tracks current position/state              |
| Cannot give values directly          | Gives values using `next()`                |
| Example: list, tuple, string, set    | Example: object returned by `iter()`       |



```python
l = [1,2,3]      # iterable
it = iter(l)     # iterator
```
