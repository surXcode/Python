#  Anonymous Function: 

- Anonymus function are use to define in the function in a single line by the help of lambda keyword.

- It is use to perform simple operation.

- It takes n number of argument but gives single output.

- Simple if else programms can be done by anonymus function.

ðŸ”¹ Syntax

- var = lambda arguments: expression
    * var(value)

```python
# simple operations and if-else programmes

add = lambda a, b: a + b
# add(10,20)
print(add(10,20)) # lambda always returns the output/value
```

```python
# square

list(map(lambda x: x * x, [1, 2, 3, 4]))
```

```python
square = lambda a:a * a
cube = lambda a: a * a * a
reverse = lambda a: str(a)[::-1]
```

```python
print(square(10))
print(cube(10))
print(reverse(1099701))
```

```python
# odd even
odd_even = lambda n: n % 2 == 0
print(odd_even(97))
```

```python
# odd even
odd_even = lambda n:"Even" if n % 2 == 0 else"odd"
print(odd_even(97))
```

```python
# greatest number among two number

greatest = lambda a, b: a if a > b else b
print(greatest(100, 25))

```

```python
# is_vowel()

vowel = lambda a: "vowel" if a.lower() in "aeiou" else "Not"

up = lambda a: "True" if a.isupper() else "False"

lo = lambda a: "True" if a.islower() else "false"

palin = lambda a: "True" if str(a) == str(a)[::-1] else "false"

positive = lambda a: "Positive" if a >= 0 else "false"
```

```python
print(vowel('A'))
print(up('Z'))
print(lo('z'))
print(palin(121))
print(positive(-5))
```

# Filter Function :

- It is unbuilt function which is use to traverse through the items of the collection and returns specific values.

- It is use to process and filter the iterable.

- The length of output may be equal or less than the length of the input.

- Filter function return filter object hence typecasting is required.

# Then takes two argument:

  - func_name
  - iterable

- function which is  to be used in filter() must returns boolean output.

# Syntax:

  - var = filter(func_name,iterable)
  - print(list(var))



```python
l1 = ['hii', 10, 20, 'Bye']


def int_val(n):
    return type(n) == int


result = filter(int_val, eval(input("enter a data:")))
print(list(result))
```

```python
# Extract the digits from string

s = "a1b2c3d4e5"

digits = list(filter(str.isdigit, s))
print(digits)
```

```python
# extract the vowel from word

word = "Programming"

vowels = list(filter(lambda ch: ch.lower() in 'aeiou', word))
print(vowels)
```

```python
# extract the odd values from a list

nums = [10, 15, 22, 33, 41, 50, 63]

odd_nums = list(filter(lambda x: x % 2 != 0, nums))
print(odd_nums)
```

```python
# extract palindrome numbers up to n.

n = 200

palindromes = list(
    filter(lambda x: str(x) == str(x)[::-1], range(1, n + 1))
)

print(palindromes)

```

```python
import math 
l1 = [3, 'hii', 4, 3.44, 'Bye', 5]
ints = filter(lambda x: type(x) == int, l1)
map = list(map(math.factorial, ints))

print(map)
```

# Reduce Function:

- It is use to reduce our iterable in it single value

- Reduce function is present in 'functools' module.

- It uses a mathematical folding technique.

# Takes two argument:

 - func_name
 - iterable

# Syntax:

   - from functools import reduce
   - var = reduce (func_name ,iterable)
   - print(var)

```python
from functools import reduce

def add(a, b):
    return a + b
result= reduce(add,[10,20,30,40])
print(result)
```

```python
from functools import reduce


def add(a, b):
    return a + b
add_val = lambda a,b : a+b
result= reduce(add_val,[1,2,3,4])
print(result)
```

```python
# factorial of a number
n = int(input("enter a number:"))

fact = lambda a, b: a * b
print(reduce (fact,range(1,n+1)))
```
