# Set Comprehension ðŸ‘
* var={var for var in collection}   --> simple looping statement
* var={var for var in collection if <condition>}  --> 1 statement block
* var={TSB if condition else FSB for var in collection}     --> 2 statement block
* var = {val fro var1 in condition for var2 in collection}   ---> nested loop

```python
# set comprehension

# extract all the unique numbers from a list.

l1 = [10, 20, 10, 50, 60]
print({i for i in l1})
```

```python
# # extract all the unique EVEN numbers from a list.

l1 = [10, 20, 10, 50, 60, 20, 55, 40, 20, 30, 60]
print({i for i in l1 if i%2==0})
```

```python
# print "positive","negative","zero" replacing the number in list

l1 = [1, -2, 0, 5, -8]
out = ["Positive" if i > 0 else "Negative" if i < 0 else "Zero" for i in l1]
print(out)
```
