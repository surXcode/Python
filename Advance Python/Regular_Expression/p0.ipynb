# Regular Expression : 

## üîç What is a Regular Expression (Regex) in Python?

- A Regular Expression (often called regex) is a powerful pattern-matching technique used to search, match, extract, or replace text in strings.
- In Python, regex is handled using the built-in re module üöÄ.
- It is a concept by the help of which can search and match a pattern in a string.
- It is a concept by the help of which can search and match a pattern in a string by using some expression is known as regular expression.


**Syntax**
```python
re.method_name("expression",string)
```


## üß† Why Use Regular Expressions?

Regex helps you:

- üîé Search for specific patterns in text

- ‚úÇÔ∏è Extract useful data (emails, phone numbers, IDs)

- üîÅ Replace or validate strings

- üß™ Validate inputs (passwords, usernames, formats)



## üì¶ Python Regex Module

- Python provides the re module for regex operations:

```python
import re
```

## üîß Common Regex Functions in Python

| Function       | Description                             |
| -------------- | --------------------------------------- |
| `re.match()`   | Matches pattern at **start** of string  |
| `re.search()`  | Searches pattern **anywhere** in string  first match|
| `re.findall()` | Returns **all matches**                 |
| `re.sub()`     | Replaces pattern with new text          |
| `re.split()`   | Splits string by pattern                |
| `re.fullmatch()` | Match entire string    |
| `re.finditer()`  | Iterator with position |
| `re.subn()`      | Replace + count        |
| `re.compile()`   | Pre-compile regex      |
| `re.escape()`    | Escape special chars   |



````python
import re

text = "I love Python"
result = re.search("Python", text)

print(result.group())
````

## üî§ Common Regex Patterns

| Pattern  | Meaning                         |
| -------- | ------------------------------- |
| `.`      | Any single character  |
| `^`      |  It matches at Start of string                 |
| `$`      | It matches at End of string                   |
| `*`      | 0 or more occurrences           |
| `+`      | 1 or more occurrences           |
| `?`      | 0 or 1 occurrence               |
| `\d`     | Digit (0‚Äì9)                     |
| `\w`     | Alphanumeric (a‚Äìz, A‚ÄìZ, 0‚Äì9, _) |
| `\s`     | Whitespace                      |
| `[abc]`  | Any one of a, b, c              |
| `[^abc]` | NOT a, b, or c                  |



## üß† Key Points to Remember

- Regex is case-sensitive by default
- Use raw strings (r"") to avoid escape issues
- Extremely useful in data cleaning, validation, and automation

````python
# üìå Example: Find All Numbers
import re

text = "My roll number is 123 and code is 456"
numbers = re.findall(r"\d+", text)

print(numbers)
````

````python
# üîÑ Example: Replace Text
import re

text = "Hello World"
new_text = re.sub("World", "Python", text)

print(new_text)
````

````python
# ‚úÖ Example: Validate Email
import re

email = "test@gmail.com"
pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
````

## üîπ 1. . ‚Üí Any Single Character

- üìå Matches exactly ONE character (except newline \n)
- üß† Think of . as: ‚Äúanything here‚Äù

Example

````python
print(re.findall("h.t", "hat hot hit hut"))
````

## üîπ 2. ^ ‚Üí Start of String

- üìå Ensures the pattern starts at the beginning
- üß† Think of ^ as: start anchor

````python
print(re.findall("^Hello", "Hello World"))
````

-‚úÖ Matches only if string starts with "Hello"

‚ùå Won‚Äôt match:
```
"Say Hello"


```

## üîπ 3. $ ‚Üí End of String

- üìå Ensures the pattern ends at the end
- üß† Think of $ as: end anchor

````python
print(re.findall("World$", "Hello World"))
````

## üîπ 4. * ‚Üí 0 or More Occurrences

- üìå Matches zero, one, or many times
- üß† b* ‚Üí b may appear or may not

````python
print(re.findall("ab*", "a ab abb abbb"))
````

## üîπ 5. + ‚Üí 1 or More Occurrences

- üìå Must appear at least once
- üß† + means mandatory repetition

````python
print(re.findall("ab+", "a ab abb abbb"))
#‚ùå "a" not included
````

## üîπ 6. ? ‚Üí 0 or 1 Occurrence

- üìå Optional ‚Äî appears once or not at all
- üß† Perfect for optional characters

Example


````python
print(re.findall("colou?r", "color colour"))
````

## üîπ 7. \d ‚Üí Digit (0‚Äì9)

- üìå Matches any numeric digit
- üß† Same as [0-9]

````python
print(re.findall("\d", "Order 123, Code 45"))
````

**‚úîÔ∏è Use a raw string (r"")**

````python
print(re.findall(r"\d", "Order 123, Code 45"))
````

**‚ö†Ô∏è Alternative (NOT recommended)**

This works because:

- \\ ‚Üí Python converts to \

- Regex then sees \d

But ‚ùå this is ugly and harder to read, especially in complex patterns.

````python
print(re.findall("\\d", "Order 123, Code 45"))
````

## üî¢ {n} in Regular Expression (Regex)
- Exactly n times
- - Matches the preceding character or group exactly n times

- üìå It is called a quantifier

**üîç Example 1: Exact Digits**

````python
text = "PIN: 1234"
print(re.findall(r"\d{4}", text))
````

**üß† \d{4} ‚Üí exactly 4 digits**

---

**Example 2: Repeating Characters**

````python
print(re.findall(r"a{3}", "aa aaaa aaa a"))
````

**üß† Matches only 3 a‚Äôs together**

--- 

**üîπ Example 3: Fixed-Length Word**


````python
print(re.findall(r"\w{5}", "Hello Python World"))
````

### üî• Related Quantifiers (IMPORTANT)

| Quantifier | Meaning               |
| ---------- | --------------------- |
| `{n}`      | Exactly n times       |
| `{n,}`     | At least n times      |
| `{n,m}`    | Between n and m times |


**üîπ {n,} ‚Üí At Least n Times**

````python
print(re.findall(r"\d{2,}", "Age 9, Code 12345"))
````

**üîπ {n,m} ‚Üí Between n and m Times**

````python
print(re.findall(r"\d{2,4}", "Years 1 12 123 12345"))
````

**üß† With Groups (Very Powerful)**
- (ab){3} ‚Üí ab repeated 3 times

````python
print(re.findall(r"(ab){3}", "ababab ab ababab"))
````

````python
print(re.findall(r"(a){2,3}", "ababab aab abaaabab"))
````

**üß™ Real-World Use Cases**

| Pattern          | Use             |
| ---------------- | --------------- |
| `\d{6}`          | OTP             |
| `\d{10}`         | Mobile number   |
| `[A-Z]{2}`       | State code      |
| `[a-zA-Z0-9]{8}` | Password length |


## üîπ 8. \w ‚Üí Word Character

- üìå Matches:

  - a‚Äìz
  - A‚ÄìZ
  - 0‚Äì9
  - _
- üß† Used for identifiers, usernames


Example

````python
print(re.findall("\w", "Hello_123 World!"))
````

````python
print(re.findall("\\w", "Hello_123 World!"))
````

````python
print(re.findall(r"\w", "Hello_123 World!"))
````

````python
print(re.findall(r"\w+", "Hello_123 World!"))
````

## üîπ 9. \s ‚Üí Whitespace

- üìå Matches:

    - Space
    - Tab
    - Newline
- üß† Used to detect or remove spaces

Example

````python
print(re.findall("\s", "Hi there"))
````

````python

print(re.findall("\\s", "Hi there"))
````

````python
print(re.findall(r"\s", "Hi there"))
````

## üîπ 10. [abc] ‚Üí Any One of a, b, or c

- üìå Matches only ONE character from set
- üß† Acts like OR

Example

````python
print(re.findall("[abc]", "apple banana cherry"))
````

## üîπ 11. [^abc] ‚Üí NOT a, b, or c

- üìå Matches any character except a, b, or c
- üß† ^ inside [] means negation

Example

````python
print(re.findall("[^abc]", "abcXYZ"))
````
