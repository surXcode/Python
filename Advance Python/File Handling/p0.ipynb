# File Handling:

- File handling lets your Python program create, read, write, append, and manage files stored on disk.

- Instead of printing data on screen, you store it permanently in files.

#  File Modes (Types)

Python provides different file modes to control how a file is opened and used.

| Mode | Meaning | Use |
|------|---------|-----|
| `r` | Read | Default mode, file must exist |
| `w` | Write | Creates new file / overwrites existing file |
| `a` | Append | Adds data at the end of the file |
| `x` | Create | Creates a new file, error if file exists |
| `b` | Binary | Used for binary files (images, pdf, etc.) |
| `t` | Text | Default text mode |
| `r+` | Read + Write | Read and write without deleting content |
| `w+` | Write + Read | Overwrites file, then allows reading |
| `a+` | Append + Read | Add data and then read |

---




```python
# Write Mode
f = open("data.txt", "w")
f.write("Hello")
f.close()
```

```python
# Append Mode
f = open("data.txt", "a")
f.write("New Line")
f.close()
```

```python
with open("data.txt", "r") as f:
    print(f.read())
```

```python
# Write names
with open("students.txt", "w") as f:
    f.write("Aman\nRiya\nSuraj")

# Append a name
with open("students.txt", "a") as f:
    f.write("\nNeha")

# Read all names
with open("students.txt", "r") as f:
    for name in f:
        print(name.strip())
```
