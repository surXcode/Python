# dictionary comprehension:

* it is use to combine two or more attribute into a single attribute with respect to the index position.

# Synatx:

* var=zip(iterable1,iterable2,...........,iterablen)

* it combines the values  and store in the form of tuple.

* zip functions return zip object

* attribute having shortest length will be consider.

```python
# for interview question

l1 = [10, 20, 30, 40]
l2 = [2.33, 3.44, 44, 44.5,55,77]
l3 = ["hii", "hello", "bye"]
data = zip(l3, l2)
for i ,j in zip(l1,l2):
    print(i,j)
# print(list(data))
```

```python
# create a dictionary by combining l1 and l2.

d1 = zip(l1, l3)
print(dict(d1))
```

# Zip comprehension:

* zip + list comprehension is a very common interview combo.

* zip() pairs elements from multiple iterables, and comprehension lets you process them in one line.

# Basic Syntax
* [(x, y) for x, y in zip(list1, list2)]

```python
# n natural number are the keys and square of numbers are values

n = int(input("Enter a number :"))
d1 = {i: i**2 for i in range(1, n + 1)}
print(d1)
```

```python
# n natural number are the keys and square of numbers are values if numbers are even.

n = int(input("Enter number: "))

result = {i: i*i for i in range(1, n+1) if i % 2 == 0}

print(result)
```

```python
# map to lists into a dictionary

l1 = [10, 20, 30]
l2 = ['hii', 'hello', 'bye']
print({i:j for i,j in zip(l1,l2)})
```

```python
# n natural numbers are the keys , square of even numbers and cube of odd numbers.

n = int(input("Enter n: "))

result = {i: (i*i if i % 2 == 0 else i**3) for i in range(1, n+1)}

print(result)
```
