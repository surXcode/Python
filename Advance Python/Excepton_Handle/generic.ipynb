## what is generic exception handling?

- Generic exception handling means catching any type of exception using a general (non-specific) exception block instead of handling different exceptions separately.
- when we do not know exact type of exception then we use generic exception.
- here we use `Exception` class to handle all the exception.
- Inthis we can not provide specific solution but can provide generic solution. 



---

## Why is it used?

Generic exception handling is useful when:
- You donâ€™t know all possible errors in advance
- You want a fallback safety net
- You want to prevent program crashes

## Drawback (Important)

## Final Takeaway (Plain Truth)

- Generic exception handling = broad safety net
- Useful for stability
- Dangerous if overused because it can hide bugs
- Best used as a last line of defense, not your first choice

```python
#Specific exception handling example
try:
    x = int(input("Enter a number: "))
    y = 10 / x
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

```python
#generic exception handling example
try:
    x = int(input("Enter a number: "))
    y = 10 / x
except Exception as e:
    print("An error occurred:", e)
else :
    print("Result is:", y)
    print("No exceptions occurred.")
finally:
    print("Execution completed.")
```

**In Generic Exception Handling, we catch all types of exceptions using the parent class Exception.**

**Meaning:**

We handle any kind of error in a single block.

## Use Case:

- Used as a safety net
- Useful in production to prevent crashes

## Disadvantage:

Can hide real bugs if overused
