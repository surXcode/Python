- By extending exception class we can create our user define exception.


**Syntax :**

```python
class Exception_name(exception):
    pass

```

```python
class Exception_name(exception):
    def __init__(self,arg1,arg2):
        super().__init__(arg1,arg2)
```


## Raise üëç

- It is used to throw the error according to user.
- by using this `raise` keyword user can throw any error inside the code.

**raise**
```
raise ExceptionName("Message")
```

ExceptionName should be standard.

````python
print("Hello from user-defined exception handling file!")
````

````python
age = int(input("Please enter your age: "))
if age < 18:
    raise ValueError("You must be at least 18 years old to proceed.")
print("You are eligible to proceed!")
````

````python
class invalidAgeError(Exception):
    """Custom exception for invalid age input."""
    pass
age = int(input("Please enter your age: "))
if age < 18:
    raise invalidAgeError("You must be at least 18 years old to proceed.")
print("You are eligible to proceed!")
````
