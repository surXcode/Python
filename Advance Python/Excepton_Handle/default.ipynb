## Default Exception Handling

- By the help of Default Exception Handling all exception including KeyboardInterrupt Error.
- Default Exception Handling refers to the automatic way a programming language handles an exception when you do NOT write your own try-except (or try-catch) block.

**In simple words:**

- If you do not handle an exception yourself, the language‚Äôs runtime system handles it by default.

## üîπ What happens in Default Exception Handling?

- When an error occurs and you don‚Äôt handle it, this happens automatically:

1. Program execution stops
2. An exception object is created
3. The runtime prints an error message (traceback)
4. The program terminates abnormally (crashes) ‚ùå

## üîπ How is this different from Custom Exception Handling?

| Default Exception Handling   | Custom Exception Handling      |
| ---------------------------- | ------------------------------ |
| Done automatically by system | Done manually by programmer    |
| Crashes the program          | Prevents crash                 |
| Shows technical error        | Can show user-friendly message |
| No recovery mechanism        | Can recover and continue       |


## üîπ Why does Default Exception Handling exist?

It ensures that:
- Errors are visible
- Bugs are not silently ignored
- Developers can debug using traceback

In other words, it‚Äôs a built-in safety and debugging mechanism.

*Note: Default Exception Handling is the automatic handling of runtime errors by the programming language when the programmer does not provide explicit exception handling.*


*Note: when we know the exact type of exception then use `specific exception` handling otherwise/else we use `Default Exception` Handling.*
