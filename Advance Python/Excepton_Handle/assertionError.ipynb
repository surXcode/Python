- It is a process of throwing error assert keyword.

- `AssertionError` occurs when an `assert` statement fails in Python.

**Syntax**
```python 
assert condition, "Error message"
```

## âœ… What is assert?

- `assert` is used for debugging â€” to check if a condition is true.
- If the condition is False, Python raises an AssertionError.

````python
x = 5
assert x == 10
````

````python
x = 5
assert x > 10, "x should be greater than 10"
````

## âœ… Where is it used?

Common use cases:

- Testing assumptions in your program
- Debugging logic errors
- Validating inputs during development

Example in a function:

````python
def divide(a, b):
    assert b != 0, "Denominator cannot be zero"
    return a / b

divide(10, 0)
````

*So donâ€™t use assert for user validation in production â€” use if and raise exceptions instead.*

````python
def check_age():
    age=int(input("Enter your age: "))
    assert age >= 18, "Not elligible to enter"
    print(f"Your age is {age}") 
try:
    check_age()
except AssertionError as e:
    print(e)
finally:
    print("Execution completed.")
````
