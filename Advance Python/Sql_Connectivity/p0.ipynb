## âœ… What is SQLite?

- SQLite is a lightweight, file-based database.
- ðŸ‘‰ No server needed. Your database is just a .db file.
- Inbuilt library by python.

## ðŸ”¹ Method: SQLite Connectivity in Python
**Step 1 â€” Import sqlite3 module**
```python
import sqlite3
```

**Step 2 â€” Create / Connect to Database**

- If the file doesnâ€™t exist, SQLite will automatically create it.
```python
conn = sqlite3.connect("mydatabase.db")
```

**Step 3 â€” Create a Cursor**

- Cursor is used to execute SQL queries.
- points to a particular location in database

```python
cursor = conn.cursor()
```

**Step 4 â€” Create a Table**
```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")
```

**Step 5 â€” Insert Data**
```python
cursor.execute("INSERT INTO students (name, age) VALUES (?, ?)", ("Suyash", 21))
conn.commit()   # Save changes
```

**Step 6 â€” Fetch Data**
```python
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for row in rows:
    print(row)

```

**Step 7 â€” Close Connection (Important)**
```python
conn.close()
```
