# Generator expression  it is use to generator object
- Syntax  
var = (i for in itereable)

by using next function we can fetch values from generator object

next() -> fetch the values at a time

Syntax 

print(next(var))

print(next(var))

```python
s = (i for i in range(1,10))

print(type(s))
print(next(s))
print(next(s))
print(next(s))
print(list(s))
print(tuple(s))
```

```python
s = (i for i in range(1,11))

print(next(s))
print(next(s))
t1 = list(s)[:-3]
print(t1)
print(tuple(s))
```
