# List Comprehension:

- List Comprehension in Python is a short, clean way to create a list using a single line of code.

- Instead of writing a for loop with append(), you write everything in one expression.

# âœ… Basic Syntax:

- [expression for item in iterable if condition]


- expression â†’ what you want to store in the list

- item â†’ variable

- iterable â†’ list, range, string, etc.

- condition â†’ optional filter in Python is a short, clean way to create a list using a single line of code.

- Instead of writing a for loop with append(), you write everything in one expression.



```python
# ðŸ”¹ Example 1 â€” Without List Comprehension
squares = []
for i in range(1,6):
    squares.append(i*i)

print(squares)
```

```python
# ðŸ”¹ Example 2 â€” With List Comprehension
squares = [i*i for i in range(1,6)]
print(squares)
```

```python
# ðŸ”¹ Example 3 â€” With Condition (Only Even Numbers)
even = [i for i in range(1,11) if i % 2 == 0]
print(even)
```

```python
# ðŸ”¹ Example 4 â€” With String
chars = [ch for ch in "python"]
print(chars)
```

```python
# ðŸ”¹ Example 5 â€” Convert Celsius to Fahrenheit
celsius = [0, 10, 20, 30]
fahrenheit = [(c * 9/5) + 32 for c in celsius]
print(fahrenheit)
```
